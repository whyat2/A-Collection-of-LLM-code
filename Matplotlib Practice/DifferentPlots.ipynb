```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#scatter plot
x = np.random.random(50) * 100
y = np.random.random(50) * 100
t = np.random.random(50) * 100 #random scalar value associated with it
#plt.scatter(x, y, c="red", s=100, marker="^", alpha=.4) #no scalar associated just plot points
plt.scatter(x, y, c=t, cmap="" s=100, marker="^", alpha=.4, vmin=10.0, vmax=90.0)
plt.show()
```

```python
#line plot
x = [1, 2, 3, 4, 5, 6, 7, 8]
x = np.array(x)
y = np.random.random(8) * 20
y2 = np.random.random(8) * 20
plt.plot(x, y, c="#0000ff", lw=3, linestyle="--")
plt.plot(x, y2, c="#00ff00", lw=3, linestyle="--")
```

```python
#bar graphs
x = ["one", "two", "three", "four"]
y = np.random.random(4) * 20
plt.bar(x, y, color="purple", align="edge", width=.8, edgecolor="red", lw=4)
plt.show
```

```python
#histogram
x = np.random.normal(20.0, 2.0, 100)
plt.hist(x, bins=[x.min(), 17, 19, 21, 23, x.max()], cumulative=True, color="blue", histtype='stepfilled')
plt.show()
```

```python
#pie chart
labels = ["A", "B", "C", "D", "F"]
x = np.random.random(5) * 40
explode = [.1, 0, 0, 0, 0]
plt.pie(x, labels=labels, explode=explode, autopct="%.2f%%", pctdistance=.6, labeldistance=1.1, startangle=90)
plt.show()
```
